```python
import pandas as pd
```

```python
single_solutions_results = pd.read_csv("../results/second-delivery/single_solution.csv")
intensification_two_opt = pd.read_csv(
    "../results/third-delivery/two_opt_intensification.csv"
)
```

```python
merged = pd.merge(
    single_solutions_results,
    intensification_two_opt,
    on="Instance",
    suffixes=("_single", "_two_opt"),
)
```

```python
merged.columns
```

```python
merged["intensification_improvement_percentage"] = (
    (merged["single_solution_cost"] - merged["local_optimum_solution_cost"])
    / merged["single_solution_cost"]
    * 100
)

print(merged["intensification_improvement_percentage"].mean())
```

```python
merged_tuned = merged.rename(
    columns={
        "single_solution_cost": "ss_cost",
        "vehicle_number_obtained_single": "ss_R",
        "local_optimum_solution_cost": "i_cost",
        "intensification_improvement_percentage": "delta_cost",
        "time_two_opt": "i_time",
    }
)[
    [
        "Instance",
        "ss_R",
        "ss_cost",
        "i_cost",
        "delta_cost",
        "i_time",
    ]
]

merged_tuned
```
