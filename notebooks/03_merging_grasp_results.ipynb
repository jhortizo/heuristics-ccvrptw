```python
import pandas as pd
```

```python
intensification_two_opt = pd.read_csv(
    "../results/third-delivery/two_opt_intensification.csv"
)
grasp_results = pd.read_csv("../results/fourth-delivery/grasp.csv")
```

```python
merged = pd.merge(
    grasp_results,
    intensification_two_opt,
    on="Instance",
    suffixes=("_grasp", "_two_opt"),
)
```

```python
merged.columns
```

```python
merged["grasp_improvement_percentage"] = (
    (merged["local_optimum_solution_cost"] - merged["grasp_optimum_solution_cost"])
    / merged["local_optimum_solution_cost"]
    * 100
)

print(merged["grasp_improvement_percentage"].mean())
```

```python
merged_tuned = merged.rename(
    columns={
        "grasp_optimum_solution_cost": "g_cost",
        "vehicle_number_obtained_grasp": "ss_R",
        "local_optimum_solution_cost": "i_cost",
        "grasp_improvement_percentage": "delta_cost",
        "time_grasp": "g_time",
    }
)[
    [
        "Instance",
        "ss_R",
        "i_cost",
        "g_cost",
        "delta_cost",
        "g_time",
    ]
]

merged_tuned
```
